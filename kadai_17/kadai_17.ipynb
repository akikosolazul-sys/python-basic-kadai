```python
class Human:
  def __init__(self,name:str,age:int):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20 :
      print(self.name,"の年齢は",self.age,"才です。大人です。")
    else:
      print(self.name,"の年齢は",self.age,"才です。大人ではありません。")


array_name = ["あつこ","さつき","りょうた","mary","Tomas"]
array_age = [20,18,3,50,60]

for i in range(5):
  h=Human(array_name[i],array_age[i])
  h.check_adult()
```
